```python
import pandas as pd
import numpy as np
import json
import os
import glob
```

```python
##############Summary##############
#This file is creating a df for 'Total number of factories in China'
#Source: https://openapparel.org/

#Total no of count
#china 1 - 12: 500
#china 13: 318
#china 14: 13
#Total=6331
```

### Import raw files

```python
df1=pd.concat(map(pd.read_json, glob.glob(os.path.join('', 'Raw Data/China Json/china*'))),ignore_index=True)
print(df1.shape)
df1.head()
```

```python
df1['features'][0]
```

### Handling features column

```python
#https://stackoverflow.com/questions/49081097/slice-pandas-dataframe-json-column-into-columns

features_expanded = df1['features'].apply(lambda x:pd.Series(x))
features_expanded.columns = ['{}.{}'.format('features',i) for i in features_expanded]
features_expanded.head()
```

### Handling features.geometry column

```python
geometry_expanded = features_expanded['features.geometry'].apply(lambda x:pd.Series(x))
geometry_expanded.columns = ['{}.{}'.format('geometry',i) for i in geometry_expanded]
geometry_expanded.head()
```

### Handling features.properties column

```python
properties_expanded = features_expanded['features.properties'].apply(lambda x:pd.Series(x))
properties_expanded.columns = ['{}.{}'.format('properties',i) for i in properties_expanded]
properties_expanded.head()
```

### Concat files

```python
final = pd.concat([df1[['type','count','next','previous']],features_expanded[['features.id','features.type']],geometry_expanded,properties_expanded],axis=1)
print(final.shape)
final.head()
```

```python
##############Unique column##############
#Check unique value for each column
def unique_counts(df):
    for i in df.columns:
        if i != 'geometry.coordinates':
            count = df[i].nunique()
            null = df[i].isnull().sum()
            print(i, "-", "unique count : ", count, ", blank count : ", null, "(", '{:,.2%}'.format(null/ df.shape[0]), ")")
            if null/ df.shape[0] > 0.2:
                print('==========Blank % > 20%==========', '\n')
            
unique_counts(final)
```

```python
final['lat']=final['geometry.coordinates'].str[1]
final['lng']=final['geometry.coordinates'].str[0]

final=final[['properties.name','properties.address','lat','lng']] #'features.id','properties.oar_id','geometry.coordinates'
final=final.rename(columns={'properties.name': 'name', 'properties.address': 'address'})
#final['name']=final['name'].str.lower()

#check for duplicate
final[final.duplicated(keep=False)].sort_values(by='name')
```

```python
#Check for no of records
total_record=df1.shape[0]
duplicated_count=final[final.duplicated(keep=False)].shape[0]
duplicated_record_unique_count=len(final[final.duplicated(keep=False)]['name'].unique())
print('Total Number of records should be :', total_record - duplicated_count + duplicated_record_unique_count)
```

```python
#Drop duplicate
final=final.drop_duplicates()
print(final.shape)

final.to_excel('Raw Data/China Json/openapparel-china.xlsx', index=False)
final.head()
```
